```python
import numpy as np
import pandas as pd
import seaborn as sns

from scipy.stats import rankdata
```

### Ainda falando sobre correlação...

```python
# Exemplos de ranking (Spearman)
x_vals = np.arange(0,10,0.1)
y_vals = np.exp(x_vals)

df = pd.DataFrame({
    'x'     : x_vals,
    'y'     : y_vals,
    'rank_x': rankdata(x_vals),
    'rank_y': rankdata(y_vals)
})

sns.scatterplot(data=df, x='rank_x', y='rank_y')
```

```python
df.corr()
```
