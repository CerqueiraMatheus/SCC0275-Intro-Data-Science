# Questão 2

## a) Comparação entre acurácia e AUC

```python
# Importa as métricas do exercício
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
# Valores de teste
y_test = train_df['Class'].values

# Arrays de teste de predição para randômico, uns e zeros
y_pred_rand = np.random.randint(2, size=y_test.shape)
y_pred_ones = np.ones(y_test.shape)
y_pred_zero = np.zeros(y_test.shape)
```

```python
# Retorna a métrica de acurácia
print(accuracy_score(y_true=y_test, y_pred=y_pred_rand))
print(accuracy_score(y_true=y_test, y_pred=y_pred_ones))
print(accuracy_score(y_true=y_test, y_pred=y_pred_zero))
```

```python
# Retorna a métrica de AUC
print(roc_auc_score(y_true=y_test, y_score=y_pred_rand))
print(roc_auc_score(y_true=y_test, y_score=y_pred_ones))
print(roc_auc_score(y_true=y_test, y_score=y_pred_zero))
```

O modelo de acurácia é mais preciso quando testado com o array de 0, mas é extremamente impreciso ao usar o array de 1. Para o array de randômicos, mantém-se na casa dos 0.5. Já o AUC também se mantém quando usado sob o array randômico, mas faz previsões estáveis e iguais a 0.5, tanto para o array de 0 quanto para o array de 1.

## b) Métrica ideal para o problema

É evidente, para o problema, uma vez que o modelo de AUC é construído para classificação binária e a variável de resposta do dataset é binária, além do fato da baixa eficiência da acurácia, que o modelo de AUC é a métrica ideal para o problema.

# Questão 3

```python
# Importa o RandomForestClassifier e cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
# Separa a coluna de resultados do treino
X = train_df.loc[:, train_df.columns != 'Class']
y = train_df['Class']

# Define o array de estimators
n_estimators = [10, 50, 100, 200]

# Define o array de max_depth
max_depth = [2, 3, 4, 5]
```

## a) 3-fold cross-validation com diferentes parâmetros num RandomForest para classificação

```python
# DF para resultados
cv_rf_1 = pd.DataFrame({
    'n_estimators': [],
    'max_depth': [],
    'auc_scores': [],
    'auc_mean': [],
    'auc_std': []
    })

# Percorre os parâmetros
for n_estimator in n_estimators:
    for cur_max_depth in max_depth:

        # Define o RandomForestClassifier
        rfc = RandomForestClassifier(
            random_state=42,
            n_estimators=n_estimator,
            max_depth=cur_max_depth,
            )
        
        # Executa o cross_validate
        crv = cross_val_score(rfc, X, y, cv=3, scoring='roc_auc', n_jobs=-1)

        # Armazena o dado
        cv_rf_1 = cv_rf_1.append({
            'n_estimators': n_estimator,
            'max_depth': cur_max_depth,
            'auc_scores': crv,
            'auc_mean': np.round(crv.mean(), 3),
            'auc_std': np.round(crv.std() * 2, 3)
            }, ignore_index=True)

# Exibe o DF gerado
cv_rf_1
```

Os resultados obtidos foram satisfatórios, organizados com AUC arredondada em três casas decimais no DataFrame.

## b) 3-fold cross-validation com diferentes parâmetros num RandomForest para classificação com *class_weight='balanced'*

```python
# DF para resultados
cv_rf_2 = pd.DataFrame({
    'n_estimators': [],
    'max_depth': [],
    'auc_scores': [],
    'auc_mean': [],
    'auc_std': []
    })

# Percorre os parâmetros
for n_estimator in n_estimators:
    for cur_max_depth in max_depth:

        # Define o RandomForestClassifier
        rfc = RandomForestClassifier(
            random_state=42,
            n_estimators=n_estimator,
            max_depth=cur_max_depth,
            class_weight='balanced'
            )
        
        # Executa o cross_validate
        crv = cross_val_score(rfc, X, y, cv=3, scoring='roc_auc', n_jobs=-1)

        # Armazena o dado
        cv_rf_2 = cv_rf_2.append({
            'n_estimators': n_estimator,
            'max_depth': cur_max_depth,
            'auc_scores': crv,
            'auc_mean': np.round(crv.mean(), 3),
            'auc_std': np.round(crv.std() * 2, 3)
            }, ignore_index=True)

# Exibe o DF gerado
cv_rf_2
```

Conforme a documentação da classe [RandomForestClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html), usar o parâmetro *balanced* no *class_weight* significa usar os valores de y para automaticamente ajustar os pesos de forma inversamente proporcional às frequências das classes do problema. No geral, as acurácias obtidas foram superiores e os desvios padrão diminuídos. Os três melhores modelos, na casa dos 98% foram, respectivamente, obtidos com as combinações de estimators e max_depth de (50, 4), (50, 5) e (100, 5). Comparando os três, o custo computacional de (50, 4) é o menor e, portanto, foi escolhido como o melhor dentre os três.

# Questão 4

```python
# Cria um novo DataFrame para alteração
df_pred = test_df.copy()

# Define o RandomForestClassifier com os parâmetros ideais
rfc = RandomForestClassifier(
    random_state=42,
    n_estimators=50,
    max_depth=4,
    class_weight='balanced'
    )

# Treina o modelo
rfc.fit(train_df.loc[:, train_df.columns != 'Class'], train_df['Class'])

# Adiciona as probabilidades ao novo DF
df_pred['Class Pred'] = rfc.predict(test_df.loc[:, train_df.columns != 'Class'])
df_pred['Proba Happ'] = [el[1] for el in rfc.predict_proba(test_df.loc[:, train_df.columns != 'Class'])]

# Exibe o DF com as predições
df_pred
```

```python
# Define o df_top com o 1% de maiores probabilidades
df_top = df_pred.sort_values(by='Proba Happ', ascending=False).head(int(len(df_pred) * 0.01))

# Tamanhos das classes
tam_pos = len(df_top[df_top['Class'] == 1])
tam_neg = len(df_top[df_top['Class'] == 0])

# Define lucro e prejuízo
lucro    = tam_pos * 100
prejuizo = tam_neg * 2

# Exibe o total
print('O total foi de R$', lucro - prejuizo, 'com', tam_pos, 'fraudes evitadas e', tam_neg, 'não fraudes bloqueadas.')
```

Considerando um universo de mais de 1400 potenciais fraudes, ter 1231 falsos positivos é extremamente satisfatório uma vez que 193 fraudes foram evitadas.
